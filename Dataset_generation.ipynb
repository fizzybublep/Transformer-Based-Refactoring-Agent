```python
import networkx as nx
from networkx.drawing.nx_pydot import read_dot
```

```python
summary_graph = read_dot(r"C:\Users\bobby\Downloads\TIKA_Refactoring_agent\detailedDeps\summary.dot")
classes_graph = read_dot(r"C:\Users\bobby\Downloads\TIKA_Refactoring_agent\detailedDeps\classes.dot")
```

```python
summary_graph = nx.DiGraph(summary_graph)
classes_graph = nx.DiGraph(classes_graph)
```

```python
#inspect the structure for both graphs
print(len(summary_graph.nodes()))
print(len(summary_graph.edges()))
```

```python
print(len(classes_graph.nodes()))
print(len(classes_graph.edges()))
```

```python
#lets calculate the cycle count
cycles_summary = list(nx.simple_cycles(summary_graph))
cycles_classes = list(nx.simple_cycles(classes_graph))

cycle_count_summary = len(cycles_summary)
cycle_count_classes = len(cycles_classes)
```

```python
print(cycle_count_summary)
```

```python
print(cycle_count_classes)
```

```python
#degree centrality
deg_summary = nx.degree_centrality(summary_graph)
deg_classes = nx.degree_centrality(classes_graph)
```

```python
#aggregated degree_summray
avg_degree_summary = sum(deg_summary.values()) / len(deg_summary)
```

```python
print(avg_degree_summary)
```

```python
#Betweeness centrality
bet_summary = nx.betweenness_centrality(summary_graph)
bet_classes = nx.betweenness_centrality(classes_graph)
```

```python
print(bet_classes)
```

```python
print(bet_summary)
```

```python
#maximum betweenness centrality for summary graphs
max_bet_summary = max(bet_summary.values())
print(max_bet_summary)
```

```python
#clustering coefficient
clust_summary = nx.clustering(summary_graph.to_undirected())
clust_classes = nx.clustering(classes_graph.to_undirected())
```

```python
print(clust_summary)
```
